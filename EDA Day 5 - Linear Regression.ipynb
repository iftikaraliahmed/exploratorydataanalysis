# UNIVARIANT ANALYSIS ON TEXT

```python
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from matplotlib import pyplot as plt
```

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
#TWITTER
tweet = pd.read_excel('narendramodi_tweets.xlsx')
```

```python
tweet.head()
```

```python
sw = set(stopwords.words('english'))

sw
```

```python
text = tweet[['text']]
```

```python
text.head()
```

```python
print(text['text'][0])
```

```python
len(text)
```

```python
# Number of words in each row

text['count'] = text['text'].apply(lambda x: len(str(x).split(" ")))
```

```python
text.head()
```

```python
# Step (1 and 2) and 3 - Converting everything in lower case  &   Replacing the special characters


text['clean_text'] =  text['text'].str.lower().str.replace('[^a-z ]','')
text.head()
```

```python
#print(text['clean_text'][0])
#print(text['text_without_stop'][0])
```

```python
# Step 4 - Removing the stopwords

stop = stopwords.words('english')

stop.extend(['amp','modi','narendra','pmo']) 

def sw(text):
    text = [word for word in text.split() if word not in stop]
    return " ".join(text)

#Applying the sw on clean_text column

text['text_without_stop'] = text['clean_text'].apply(sw)

print(text['clean_text'][0])
print(text['text_without_stop'][0])
```

```python
#Starting the count vector

vec = CountVectorizer()

vec.fit(text['text_without_stop'])

#Creating the dict of item in the matrix
#dic = vec.vocabulary_.items()
#dic
```

```python
#vec = CountVectorizer(min_df = 0.2)

#vec.fit(text['text_without_stop'])
```

```python
# DTM Matrix

x = vec.fit_transform(text['text_without_stop'])

x
```

```python
# Convert this DTM into Data Frame

DTM = pd.DataFrame(x.toarray(),columns=vec.get_feature_names())

DTM.head()

#vec.get_feature_names()
```

```python
#Calculating the word frequency

word_freq = DTM.sum()

word_freq.sort_values(ascending = False).head()
```
