```python
terror_frame = [gtd_70to91,gtd_92to11_no_93,gtd1993_748,gtd_12to15]
final_data = pd.concat(terror_frame,axis=0,join='outer',sort=True)
final_data
print(type(terror_frame))
print(type(final_data))
```

# Q1. How have casualities evolved throughout the years?

```python
final_data['Total_Casualities']=final_data['nkill']+final_data['nwound']
per_year_casualities_final_data=final_data.groupby('iyear')['Total_Casualities'].sum().reset_index()
per_year_casualities_final_data
```

```python
import matplotlib.pyplot as plt
per_year_casualities_final_data.plot(x='iyear',y='Total_Casualities',color='red')
plt.rcParams['figure.figsize']=[15,6]
plt.xlabel('iyear')
plt.ylabel('Total_Casualities')
plt.title('Terror Casualities Year Wise')
plt.show()
```

# Q2. Are certain nationalities more targeted? If yes, which one?

```python
attack_nationality_wise_final_data = pd.crosstab(index=final_data.natlty1_txt,columns='count').reset_index()
attack_nationality_wise_final_data.sort_values(by='count',ascending=False)
#print(attack_nationality_wise_final_data['count'].sort_values(ascending=False))
```

Answer - Iraq is the most targeted nation

# Q3. Are some countries better at defending themselves against the terrorist attacks? If yes, then which is the safest country to live 

```python
safest_country_final_data = final_data.groupby('country_txt')['success','nkill','nwound'].sum().reset_index()
safest_country_final_data.sort_values(by='success',ascending=True).head(4)
```
