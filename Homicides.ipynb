```python
#incident family relatons
family = ['Brother','Sister','Mother','Son','Daughter','Husband','Wife','Father']
familyhomicide = homicidedata[homicidedata['Relationship'].isin(family)]
 
```

```python
def desc(df):
    if ((df['Relationship'] == 'Brother') & (df['Perpetrator Sex'] == 'Male')):
            return "Brother killed by brother"
    elif ((df['Relationship'] == 'Brother') & (df['Perpetrator Sex'] == 'Female')):
            return "Brother killed by sister"
    elif ((df['Relationship'] == 'Sister') & (df['Perpetrator Sex'] == 'Male')):
            return "Sister killed by brother"
    elif ((df['Relationship'] == 'Sister') & (df['Perpetrator Sex'] == 'Female')):
            return "Sister killed by sister"
    elif ((df['Relationship'] == 'Mother') & (df['Perpetrator Sex'] == 'Male')):
            return "Mother killed by son"
    elif ((df['Relationship'] == 'Mother') & (df['Perpetrator Sex'] == 'Female')):
            return "Mother killed by daughter"
    elif ((df['Relationship'] == 'Father') & (df['Perpetrator Sex'] == 'Male')):
            return "Father killed by son"
    elif ((df['Relationship'] == 'Father') & (df['Perpetrator Sex'] == 'Female')):
            return "Father killed by daughter"
    elif ((df['Relationship'] == 'Son') & (df['Perpetrator Sex'] == 'Male')):
            return "Son killed by father"
    elif ((df['Relationship'] == 'Son') & (df['Perpetrator Sex'] == 'Female')):
            return "Son killed by mother"
    elif ((df['Relationship'] == 'Daughter') & (df['Perpetrator Sex'] == 'Male')):
            return "Daughter killed by father"
    elif ((df['Relationship'] == 'Daughter') & (df['Perpetrator Sex'] == 'Female')):
            return "Daughter killed by mother"
    elif ((df['Relationship'] == 'Husband') & (df['Perpetrator Sex'] == 'Female')):
            return "Husband killed by wife"
    elif ((df['Relationship'] == 'Wife') & (df['Perpetrator Sex'] == 'Male')):
            return "Wife killed by husband"
    else :
        return ""
```

```python
res = pd.DataFrame()
for i,data in familyhomicide.iterrows():
    cur_row = { 
        "Record ID" : data['Record ID'],
        "desc" :desc(data)
    }
    res = res.append(cur_row,ignore_index=True)
familyhomicide = familyhomicide.merge(res)  
familyhomicide.head()
```

```python
import string
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from matplotlib import pyplot as plt
```

```python
#reviews.text
from nltk.corpus import stopwords
amazonreview['clean_txt'] = amazonreview['reviews.text'].str.lower().str.replace('[^a-z ]','')
stop = stopwords.words('english')
stop.extend(['use'])
def sw(txt):
    txt = [word for word in txt.split() if word not in stop]
    return " ".join(txt)

amazonreview['text_without_stop'] = amazonreview['clean_txt'].apply(sw)
# starting the count vector

vec = CountVectorizer()
vec.fit(amazonreview['text_without_stop'])
x = vec.fit_transform(amazonreview['text_without_stop'])
DTM  = pd.DataFrame(x.toarray(),columns=vec.get_feature_names())
DTM.sum().nlargest(5)
```
