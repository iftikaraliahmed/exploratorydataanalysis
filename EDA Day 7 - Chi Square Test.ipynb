# Bivariate analysis - Categorical vs. Categorical - Chi Sqaure Test

```python
from pydataset import data

survey = data('survey')
```

H0: Exercise habit does not depend on smoking habit
H1: Exercise habit does depend upon smoking habit

```python
# 1) Convert the column of smoke and exer column into cross table
import pandas as pd
```

```python
tble = pd.crosstab(index=survey['Smoke'],columns=survey['Exer'])
```

```python
tble
```

```python
#using chi sqaure function
from scipy.stats import chi2_contingency
```

```python
#Testing the hypothesis
chi2_contingency(tble)
```

```python
'''(5.488545890584232,--> chi square
 0.48284216946545633,--> pvalue
 6,---> degree of freedom
 array([[ 5.36016949,  1.0720339 ,  4.56779661],
        [92.09745763, 18.41949153, 78.48305085],
        [ 9.25847458,  1.85169492,  7.88983051],
        [ 8.28389831,  1.65677966,  7.05932203]]))'''
```

```python
#where pvalue is > alpha(0.05) --> accepting null hypothesis
```

```python
#if the diferrence between frequency table values and the values in the array is not much then the data is normally distributed
```

```python
# Group bar chart

#finding out mean age
tble1 = survey.groupby(['Smoke','Exer'])['Age'].mean().reset_index()
```

```python
tble1 = tble1.pivot(index='Smoke',columns='Exer',values='Age')
```

```python
tble1
```

```python
#Converting this into a group bar chart

from matplotlib import pyplot as plt

tble1.plot(kind = 'bar')

plt.rcParams['figure.figsize']=[8,6]
#Adjusting the legends

ax = plt.subplot(111)
ax.legend(loc = 'upper left',ncol = 3)



plt.show()
```

```python
#Stacked bar chart
from matplotlib import pyplot as plt


tble1.plot(kind = 'bar', stacked=True)
plt.rcParams['figure.figsize']=[10,8]
#Adjusting the legends

ax = plt.subplot(111)
ax.legend(loc = 'upper left',ncol = 3)



plt.show()
```
