```python
Association Rules - 5 step process
```

```python
#Step 1

productCountDf = trans_part1.groupby(['product_id']).size().reset_index()

productCountDf.head()

#Renaming the columns

productCountDf = productCountDf.rename(columns={ 0 : "Freq"})

#Finding top 20 products

productCountDf_top_20 = productCountDf.sort_values(by='Freq',ascending = False)
productCountDf_top_20.head(20)
```

```python
#Bringing the column of product name

newproductCountDf = pd.merge(productCountDf_top_20,product_df, left_on="product_id",right_on="product_id",how='left')
newDf=newproductCountDf[['product_id','product_name']]
newDf.head()
```

```python
#list for which i need to filter my trans_df

productId = newDf[['product_id']]
#productId
```

```python
#Filtering the trans_df for these 20 product id

df = trans_part1[0:0]

for i in range(0,19):
    pId = productId.iloc[i]['product_id']
    stDf = trans_part1[trans_part1['product_id']==pId]
    df = df.append(stDf, ignore_index = False)
df.head()
```

```python
# Getting the names for each product id 

df = pd.merge(df, newDf, left_on="product_id",right_on="product_id",how="left")
df.head()
```

```python
#Lets create the long table of order_id and product_name

basket = df.groupby(['order_id','product_name']).size().reset_index()
```

```python
basket.head()
```

```python
#Converting the long data to wide data

basket1 = basket.pivot('order_id','product_name',0)
```

```python
basket1.head()
```

```python
final_basket = basket1.fillna(0)
```

```python
final_basket.head()
```

```python
#Creating the frequently occuring itemset

freq_itemset = apriori(final_basket, min_support=0.01, use_colnames = True)
```

```python
freq_itemset.sort_values(by = 'support', ascending=False)
```

```python
rules = association_rules(freq_itemset, metric='lift',min_threshold=1.05)
rules.sort_values(by = 'lift', ascending=False)
```
